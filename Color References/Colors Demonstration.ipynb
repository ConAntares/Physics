## Colors of  Lines and Brackets



````python
# Colors of  Lines and Brackets

import matplotlib.pyplot as plt

# Define the color codes for each line
"""
Red =       [rgb(210,060,065),rgb(200,020,035),rgb(225,045,050),rgb(240,045,035),rgb(255,045,030),rgb(250,100,075)]
Orange =    [rgb(240,120,080),rgb(235,115,030),rgb(245,130,010),rgb(250,140,020),rgb(255,150,030),rgb(250,160,060)]
Yellow =    [rgb(245,175,075),rgb(225,180,000),rgb(240,190,030),rgb(250,200,030),rgb(255,210,030),rgb(250,210,060)]
Green =     [rgb(030,110,110),rgb(040,140,060),rgb(050,165,060),rgb(060,190,060),rgb(070,210,060),rgb(090,180,120)]
Blue =      [rgb(050,085,140),rgb(020,090,170),rgb(020,100,225),rgb(020,130,240),rgb(020,160,255),rgb(090,120,240)]
Violet =    [rgb(070,060,150),rgb(085,075,160),rgb(100,090,200),rgb(120,105,240),rgb(140,120,255),rgb(120,080,240)]
Purple =    [rgb(120,060,120),rgb(120,060,140),rgb(160,070,185),rgb(200,080,220),rgb(240,120,255),rgb(185,095,245)]
Wine =      [rgb(160,050,100),rgb(160,030,100),rgb(185,035,110),rgb(210,040,120),rgb(235,045,130),rgb(240,080,140)]
Brown =     [rgb(140,100,085),rgb(150,100,080),rgb(170,115,085),rgb(190,130,090),rgb(210,145,100),rgb(220,150,110)]
Grey =      [rgb(060,060,060),rgb(090,090,090),rgb(120,120,120),rgb(150,150,150),rgb(180,180,180),rgb(210,210,210)]
"""

Red =       ["#D23C41", "#C81423", "#E12D32", "#F02D23", "#FF2D1E", "#FA644B"]
Orange =    ["#F07850", "#EB730E", "#F5820A", "#FA8C14", "#FF961E", "#FAA03C"]
Yellow =    ["#F5AF4B", "#E1B400", "#F0BE1E", "#FAC81E", "#FFD21E", "#FAD23C"]
Green =     ["#1E6E6E", "#288C3C", "#32A53C", "#3CBE3C", "#46D23C", "#5AB478"]
Blue =      ["#32558C", "#145AAA", "#1464E1", "#1482F0", "#14A0FF", "#5A78F0"]
Violet =    ["#463C96", "#554BA0", "#645AC8", "#7869F0", "#8C78FF", "#7850F0"]
Purple =    ["#783C78", "#783C8C", "#A046B9", "#C850DC", "#F078FF", "#B95FF5"]
Wine =      ["#A03264", "#A01E64", "#B9236E", "#D22878", "#EB2D82", "#F0508C"]
Brown =     ["#8C6455", "#966450", "#AA7355", "#BE825A", "#D29164", "#DC966E"]
Grey =      ["#3C3C3C", "#5A5A5A", "#787878", "#969696", "#B4B4B4", "#D2D2D2"]

# Combine the color codes into a list
line_colors = [Grey, Brown, Wine, Purple, Violet, Blue, Green, Yellow, Orange, Red]

params = {"text.usetex": False, "font.family": "serif", "mathtext.fontset": "cm",
          "axes.titlesize": 16, "axes.labelsize": 14, "figure.facecolor": "w"}
plt.rcParams.update(params)

# Create a canvas
fig, ax = plt.subplots(1, 1, figsize=(10,6), dpi=196)
plt.title("Colors of Lines and Brackets")

# Iterate through the color codes and draw the corresponding color blocks
for row, color_row in enumerate(line_colors):
    for col, color in enumerate(color_row):
        # Draw a rectangle to represent the color
        rect = plt.Rectangle((col, row), 1, 1, color=color)
        ax.add_patch(rect)
        # Add the color code text in the middle of the rectangle
        plt.text(col + 0.5, row + 0.5, color, ha="center", va="center", fontsize=8, color="white")

# Set the range and labels for the axes
ax.tick_params(direction="in", which="both", top=True, right=True, bottom=True, left=True)
max_length = max(len(Red), len(Orange), len(Yellow), len(Green))
ax.set_xlim(0, max_length)
ax.set_ylim(0, len(line_colors))
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_yticks(range(len(line_colors)))

labels = ["Grey", "Brown", "Wine", "Purple", "Violet", "Blue", "Green", "Yellow", "Orange", "Red"]
yaxis_offset = 0.5  # move labels up by 0.5
for i, label in enumerate(labels):
    ax.text(-max_length*0.001, i + yaxis_offset, label, transform=ax.get_yaxis_transform(), ha="right", va="center")
````
