# Linear Dynamical Systems in Python

## Basic Linear Dynamical System
Let's simulate a time-invariant linear dynamic system  
$x_{t+1}=Ax_t, t=1,...,T$  
with dynamics matrix  
$A= \begin{bmatrix}
    0.97 &  0.10 & -0.05 \\
   -0.30 &  0.99 &  0.05 \\
    0.01 & -0.04 &  0.96
    \end{bmatrix}$  
and initial state $x_1=(1,0,-1)$. We store the state trajectory in the $n \times T$ matrix `state_traj`, with the $i$th column $x_t$. 

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.matrix([ 1, 0, -1])     # Initinal state
n  = x1.size
T  = 50 
A  = np.matrix([[ 0.97, 0.10,-0.05],
                [-0.30, 0.99, 0.05],
                [ 0.04,-0.04, 0.96]])

state_traj = np.hstack((np.transpose(x1), np.zeros((n,T-1))))

for t in np.arange(1,T):
    state_traj[:,t] = np.dot(A,state_traj[:,t-1])

print()

plt.rc("text", usetex = True)
plt.plot(np.arange(0,T), np.transpose(state_traj)[:,0], label = r"$xt_1$",)
plt.plot(np.arange(0,T), np.transpose(state_traj)[:,1], label = r"$xt_2$",)
plt.plot(np.arange(0,T), np.transpose(state_traj)[:,2], label = r"$xt_3$",)
plt.legend(loc='best')
```
