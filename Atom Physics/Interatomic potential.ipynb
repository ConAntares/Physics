# Interatomic potential

Interatomic potentials are mathematical functions to calculate the potential energy of a system of atoms with given positions in space. Interatomic potentials are widely used as the physical basis of molecular mechanics and molecular dynamics simulations in computational chemistry, computational physics and computational materials science to explain and predict materials properties. Examples of quantitative properties and qualitative phenomena that are explored with interatomic potentials include lattice parameters, surface energies, interfacial energies, adsorption, cohesion, thermal expansion, and elastic and plastic material behavior, as well as chemical reactions.

* Interatomic potential:
$$ \phi(r)=-\frac{B}{r^m}+\frac{A}{r^n} $$
$$ n>m $$

* Interatomic force:
$$ \vec{F}(\vec{r})=-\nabla\phi(\vec{r}) $$
$$ F(r) = -\frac{\rm d}{{\rm d}r}\phi(r) $$


```python
#### Interatonic potential

import numpy as np

A = 1; B = 2
m = 2; n = 3
step = 0.01
r = np.array(np.arange(step,10,step))

# Repulsive potential
rep = A/(r**n)
# Attractive potential
atp = -B/(r**m)
# Resulting potential
pot = rep + atp

import matplotlib; import matplotlib.pyplot as plt

plt.figure(dpi=192); params = {"text.usetex":True, "font.family":"serif", "mathtext.fontset":"cm", "axes.titlesize": 16, "axes.labelsize":14, "figure.facecolor":"w"}
matplotlib.rcParams.update(params)
plt.ticklabel_format(style="sci", scilimits=(0,0)); plt.tick_params(direction="in",top=True,right=True,bottom=True,left=True)
plt.title("Interatomic potential"); plt.xlabel(r"Interatomic distance"); plt.ylabel(r"Interatomic potential")
plt.xlim(0, 8); plt.ylim(-4, 4); 
plt.plot(r, rep, label="Repulsive potential", color="#FA3C3C")
plt.plot(r, atp, label="Attractive potential", color="#0A8CFF")
plt.plot(r, pot, label="Resulting potential", color="#6E64FA")
plt.plot(r, r*0, color="gray", alpha=0.4); plt.legend(loc="best")

```

```python
#### Interatonic force

import numpy as np

A = 1; B = 2; 
m = 2; n = 3; 
step = 0.01
r = np.array(np.arange(step,10,step))

# Repulsive force
ref = -n*A/(r**(n+1))
# Attractive force
atf = m*B/(r**(m+1))
# Resulting force
force = ref + atf

import matplotlib; import matplotlib.pyplot as plt

plt.figure(dpi=192); params = {"text.usetex":True, "font.family":"serif", "mathtext.fontset":"cm", "axes.titlesize": 16, "axes.labelsize":14, "figure.facecolor":"w"}
matplotlib.rcParams.update(params)
plt.ticklabel_format(style="sci", scilimits=(0,0)); plt.tick_params(direction="in",top=True,right=True,bottom=True,left=True)
plt.title("Interatomic force"); plt.xlabel(r"Interatomic distance"); plt.ylabel(r"Interatomic force")
plt.xlim(0, 8); plt.ylim(-4, 4)
plt.plot(r, ref, label="Repulsive force", color="#FA9C3D")
plt.plot(r, atf, label="Attractive force", color="#00C8F0")
plt.plot(r, force, label="Resulting force", color="#A0EB46")
plt.plot(r, r*0, color="gray", alpha=0.4); plt.legend(loc="best")
```

### Turning Atoms in Ions

Example ${\rm NaCl}$

* Ionization energy ${\rm Na}$: $5.1{\rm eV}$

* electron affinity ${\rm Cl}$: $3.6{\rm eV}$

* net energy cost: $5.1{\rm eV}-3.6{\rm eV}=1.5{\rm eV}$ per pair

### Ionic bonding

Example ${\rm NaCl}$

* Potential energy: 
$$ E_{Coulomb} = -\frac{e^2}{4\pi\epsilon_0 a} $$

* Consider the centre ${\rm Na}$ ion, energy gain from next 6 ${\rm Cl}$ ions:
$$ E = -6\frac{e^2}{4\pi\epsilon_0 a}$$

* Energy loss from next 12  ${\rm Na}$ ions:
$$ E = +12\frac{e^2}{4\pi\epsilon_0 a\sqrt{2}}$$

* Energy gain from 8 more ${\rm Cl}$ ions:
$$ E = -8\frac{e^2}{4\pi\epsilon_0 a\sqrt{3}}$$

* The total energy:
$$ E = -\frac{e^2}{4\pi\epsilon_0 a}\times(6-\frac{12}{sqrt{2}}-\frac{8}{\sqrt{3}}) $$

* For one ion, the series converges:
$$ E \approx -1.748\frac{e^2}{4\pi\epsilon_0a} = -M_d\frac{e^2}{4\pi\epsilon_0a} $$

* $M_d$ is called the madelung constant. It is specific for a given structure.

### The covalent bonding

* Electron configuration: $2s$ and $2p$ electrons;

* Formation of four $sp^3$ hybrid orbitals as linear combination between the $s$ and three $p$ orbitals;

* Directional character of $p$ orbitals is also found in $sp^3$ orbitals.

* Cohensive energies sillilar to ionic bonding, in the $\rm eV$ range.

* Very directional bonding.

### Metallic bonding

* Outer electrons are deloaclized and act as "glue" between positively charged ion cores;

* Generally found for elements with one, two, or three valence electrons;

* Cohesive energies in the $\rm eV$ range.

### More characteristics of Metallic bonding in Simple metals

* Smaller cohesive energies than in ionic crystals;

* Larger ionic radii,   
e.g. for $\rm Na$: $3.82 {\rm A}$(metal); $1.94 {\rm A}$($\rm NaCl$);

* Bonding has no directional perference;

* Closed-packed atomic configurations are preferred: best possible overlap between the orbitals.

### Metallic bonding: Why is this so favorable?

* In Schrodinger equation:
$$ i\hbar\frac{\partial}{\partial t}\psi(\vec{r},t) = -\frac{\hbar^2}{2m}\nabla^2\psi(\vec{r},t) + V(\vec{r},t)\psi(\vec{r},t) $$
The part of kinetic energy is proportional to (negative) average curvature of wave function; "Flatter" wave function means lower energy;

* For heisenberg uncertainty principle:
$$ \Delta x\Delta p \geq \hbar/2 $$
The less localization means smaller $p$ variation, and smaller kinetic energy;

### Transition metals

* $4s$ and $3d$ have very similar energies;

* $4s$ electrons form deloaclized metallic bonds;

* $3d$ electrons form more local (covalent-like) bonds;

* Higher cohesive energies.

### Van der Waals Bonding

* Also known as "Fluctuating Dipole Forces", "Dispersion Forces", or "Molecular Bonding"

For two atoms/molecules far apart, there is attraction due to Van der Waals forces. Both atoms have a dipole moment which may be zero on average, but can fluctuate momentarily.

If one atom obtains a momentary dipole moment, $p_1$ gives rise to an electric field that second atom experiences:

$$ E = \frac{p_1}{4\pi\epsilon_0 r^3} $$

The second atom can polarize, also obtaining a dipole moment $p_2$ to lower its energy.

As a result, the two atoms will attract each other.

### Van der Waals Force

$$ U = -\vec{E}\cdot\vec{p_2} = -\vec{E}\cdot\alpha\vec{E} $$
$$ \propto -\frac{1}{r^3}\alpha\frac{1}{r^3} \propto -\frac{1}{r^6} $$

* Dipole moment caused by fluctuations;

* This is always present as an attractive force (even between $\rm He$ atoms as in this case);

* It is very weak and depends on the distance as $r^{-6}$;

* Lowest melting temperature: rare-gas solids: $\rm Ne$, $\rm Ar$, $\rm Kr$, $\rm Xe$: only van der Waals Bonding, less than room temperature; then simple metals and noble metals and finally, transition metals (covalently bonded solids and ionic crystals).
