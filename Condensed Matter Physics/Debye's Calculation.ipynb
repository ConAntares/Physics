# Debye's Calculation

In 1912, Peter Debye, discovered how to better treat the quantum mechanics of oscillations of atoms, and managed to explain the $T^3$ dependance of the specific heat. Debye realized that oscillation of atoms is the same thing as sound, and sound is a wave, so it should be quantized the same way as Plank had quantized light wave in 1900. Besides the fact that the speed of light is much faster than that of sound, there is only one minor difference between light and sound: for light, there are two polarizations for each wavevector $\vec k$ for each $\vec k$, whereas for sound there are three modes for each wavevector $\vec k$. For simplicity of presentation here we will assume that the transverse and longitudinal modes have the same velocity, although in truth the longitudinal velocity is usually somewhat greater than the transverse velocity.

## Periodic (Born-von Karman) boundary conditions

We imagine that a length $L$, its two ends are connected together making the sample into a circle. The periodic boundary condition means that, any wave in this sample $e^{ikr}$ is required to have the same value for a position $r$ as it has for $r+L$. This then restricts the possible values of $k$ to be:

$$ k=\frac{2\pi n}{L} $$

for $n$ an integer. If we are ever required to sum over all possible values of $k$, for large enough $L$ we can replace the sum with an integral obtaining:

$$ \sum_{k}=\frac{L}{2\pi}\int^{\infty}_{-\infty}{\rm d}k $$

For three dimensions, for a sample of size $L^3$, we identify opposite ends of the sample. The value of wavevector $\vec k$ can be described as:

$$ \vec k=\frac{2\pi n}{L}(n_1,n_2,n_3) $$

For integer values of $n_i$ so here each $\vec k$ point now occupies a volume of $(2\pi n/L)^3$. Because of this discretization of values of $\vec k$, whenever we have a sum over all possible $\vec k$ values we obtain:

$$ \sum_{k}=\frac{L^3}{(2\pi)^3}\int^{\infty}_{-\infty}{\rm d}{\vec k} $$

## Debye's calculation following planck

Debye decided that the oscillation modes of a solid were waves with frequencies $\omega(\vec k)=v|\vec k|$ with $v$ the sound velocity, and for each $\vec k$ there should be three  possible oscillation modes, one for each direction of motion.

$$ \braket{E} = 3\sum_{\vec k}\hbar\omega(\vec k)\left(n_B(\beta\hbar\omega(\vec k))+\frac{1}{2}\right)
\\ = 3\frac{L^3}{(2\pi)^3}\int{\rm d}{\vec k} \hbar\omega(\vec k) \left(n_B(\beta\hbar\omega(\vec k))+\frac{1}{2}\right) $$

Each excitation mode is boson of frequency $\omega(\vec k)$ and is occupied on average $n_B(\beta\hbar\omega(\vec k)$ times.

By spherical symmetry, we may convert the three-dimensional integral to a one-dimensional integral:

$$ \int{\rm d}{\vec k} = 4\pi\int^{\infty}_{0}k^2{\rm d}k $$

We recall that $4\pi k^2$ is the area of the surface of a sphere of radius $k$ and use $k=\omega/v$ to obtain:

$$ \braket{E}=3\frac{4\pi L^3}{(2\pi)^3}\int^\infty_0\omega^2{\rm d}\omega(1/v^3)(\hbar\omega) \left(n_B(\beta\hbar\omega(\vec k))+\frac{1}{2}\right) $$

This result for the quantum energy of the sound waves is strikingly similar to Planck's result for the quantum energy of light waves, only we have replaced $2/c^3$ by $3/v^3$

It is convenient to replace $nL^3=N$ where $n$ is the density of atoms. We obtain:

$$ \braket{E}=\int^\infty_0{\rm d}\omega g(\omega)(\hbar\omega) \left(n_B(\beta\hbar\omega(\vec k))+\frac{1}{2}\right) $$

where the density of sates $g(\omega)$ is given by:

$$ g(\omega)=L^3\left[\frac{12\pi\omega^2}{(2\pi)^3v^3}\right] = N\left[\frac{12\pi\omega^2}{(2\pi)^3nv^3}\right] = N\frac{9\omega^2}{\omega^3_d} $$

where:

$$ \omega^3_d = 6\pi^2nv^3 $$

This frequency will be know as the Debye frequency.

The meaning of the density of sates here is that the total number of oscillation modes with frequencies between $\omega$ and $\omega+{\rm d}\omega$ is given by $g(\omega){\rm d}\omega$.

At any rate, this zero-point energy gives us a contribution which is temperature independent. Since we are concerned with $ C=\partial\braket{E}/\partial T$ this term will not contribute and we weill separate it out. We thus obtain:

$$ \braket{E} = \frac{9N\hbar}{w\omega^3_d}\int^\infty_0{\rm d}\omega\frac{\omega^3}{e^{\beta\hbar\omega}-1} + T \text{ independent constant}$$

We can define a variables $x=\beta\hbar\omega$, this becomes:

$$ \braket{E} = \frac{9N\hbar}{w\omega^3_d(\beta\hbar)^4}\int^\infty_0{\rm d}x\frac{x^3}{e^x-1} + T \text{ independent constant}$$

By calculating this nasty integral, we obtain:

$$ \braket{E} = 9N\frac{(k_B T)^4}{(\hbar\omega_d)^3}\frac{\pi^4}{15} $$

Notice the similarity to Planck's derivation of the $T^4$ energy of photons. As a result, the heat capacity is

$$ C = \frac{\partial\braket{E}}{\partial T} =  Nk_B\frac{(k_B T)^3}{(\hbar\omega_d)^3}\frac{12\pi^4}{5} \sim T^3 $$

This correctly obtains the desired $T^3$ specific heat. Furthermore, the prefactor of $T^3$ can be calculated in terms of know quantities such as the sound velocity. Note that the Debye frequency in this equation is sometimes replaced by Debye temperature:

$$ k_BT_{\rm Debye} = \hbar\omega_d $$

$$C = \frac{\partial\braket{E}}{\partial T} =  Nk_B\frac{(T)^3}{(T_{\rm Debye})^3}\frac{12\pi^4}{5} $$

## Low temperature limit

The temperature of a Debye solid is said to be low if $T \ll T_D$, leading to

$$ \frac{C_V}{Nk_B} \sim 9\left(\frac{T}{T_{\rm Debye}}\right)^3\int^\infty_0\frac{x^4e^x}{(e^x-1)^2}{\rm d}x $$

This definite integral can be evaluated exactly:

$$ \frac{C_V}{Nk_B} \sim \frac{12\pi^4}{5}\left(\frac{T}{T_{\rm Debye}}\right)^3 $$

In the low temperature limit, the limitations of the Debye model mentioned above do not apply, and it gives a correct relationship between (phononic) heat capacity, temperature, the elastic coefficients, and the volume per atom (the latter quantities being contained in the Debye temperature).
